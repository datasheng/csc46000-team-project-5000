# Historical Traffic Data

```python
df_traffic = pd.read_csv("/content/Traffic_Volume_Counts_(Historical)_20260317.csv")
```

```python
df_traffic.info()
```

```python
hour_cols = df_traffic.columns[7:31]  # adjust slice to match your actual hour columns
df_traffic[hour_cols] = df_traffic[hour_cols].astype(str)
df_traffic[hour_cols] = df_traffic[hour_cols].apply(lambda x: x.str.replace(',', '', regex=False))
df_traffic[hour_cols] = df_traffic[hour_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df_traffic.info()
```

```python
df_traffic['total_volume'] = df_traffic[hour_cols].sum(axis=1)
```

```python
df_traffic.head()
```

```python
df_daily = df_traffic.groupby('Date')['total_volume'].sum().reset_index()
```
