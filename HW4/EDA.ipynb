# Exploratory Data Analysis (EDA)
Dataset: StudentPerformanceFactors.csv

This notebook performs exploratory data analysis on a dataset of student performance factors. The goal is to understand the distribution of variables and relationships that may influence exam scores.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

## 2. Load the Dataset

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
df.head()
```

## 3. Dataset Overview

```python
df.shape
```

```python
df.info()
```

## 4. Check Missing Values

```python
df.isnull().sum()
```

## 5. Summary Statistics

```python
df.describe()
```

```python
df.describe(include='object')
```

## 6. Separate Numerical and Categorical Columns

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

print('Numerical Columns:', num_cols)
print('Categorical Columns:', cat_cols)
```

## 7. Histograms for Numerical Variables

```python
for col in num_cols:
    plt.figure()
    sns.histplot(df[col])
    plt.title(f'Histogram of {col}')
    plt.show()
```

## 8. Bar Charts for Categorical Variables

```python
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=df[col])
    plt.title(f'Bar Chart of {col}')
    plt.show()
```

## 9. Boxplot of All Numerical Variables

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df[num_cols])
plt.title('Boxplot of Numerical Variables')
plt.show()
```

## 10. Pairwise Scatter Plot Matrix (Numerical Only)

```python
sns.pairplot(df[num_cols])
plt.show()
```

## 11. Categorical vs Categorical Relationships

```python
pd.crosstab(df['Gender'], df['Motivation_Level'])
```

```python
sns.countplot(x='Motivation_Level', hue='Gender', data=df)
plt.title('Motivation Level by Gender')
plt.show()
```

## 12. Numerical vs Categorical Relationships

```python
sns.scatterplot(x='Hours_Studied', y='Exam_Score', hue='Gender', data=df)
plt.show()
```

```python
sns.boxplot(x='Motivation_Level', y='Exam_Score', data=df)
plt.show()
```

## 13. Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols].corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```
