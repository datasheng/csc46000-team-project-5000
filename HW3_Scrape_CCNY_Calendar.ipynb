```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.ccny.cuny.edu/registrar/fall'
r = requests.get(url)

html_doc = r.text

soup = BeautifulSoup(html_doc)
```

```python
table = soup.find('table')
rows = table.find_all('tr')
```

```python
dateCol = []
daysCol = []
descriptionCol = []

for row in rows:
  cols = row.find_all('td')
  if not cols:
    continue

  date = cols[0].text.strip()
  day = cols[1].text.strip()
  description = cols[2].text.strip()

  dateCol.append(date)
  daysCol.append(day)
  descriptionCol.append(description)

```

```python
df = pd.DataFrame({'Date': dateCol, 'Day': daysCol, 'Description': descriptionCol})
```
