```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import kagglehub
```

```python

path = kagglehub.dataset_download("aayushmishra1512/twitchdata")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path + "/twitchdata-update.csv")
df.head()
```
