# CUNY Fall 2021 Academic Calendar Scraper

```python
# Import required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Fetch the webpage from URL
url = 'https://www.ccny.cuny.edu/registrar/fall'
response = requests.get(url)
```

```python
# Parse the HTML with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the calendar table
table = soup.find('table')
```

```python
# Extract table rows
rows = table.find_all('tr')

# Initialize lists
dates = []
days = []
texts = []
```

```python
# Parse each row
for row in rows[1:]:
    cells = row.find_all('td')
    if len(cells) >= 2:
        # Get date and day columns
        date_text = cells[0].get_text(strip=True)
        day_text = cells[1].get_text(strip=True)

        # Get description (may be in 3rd column or combined)
        if len(cells) >= 3:
            desc_text = cells[2].get_text(strip=True)
        else:
            desc_text = ""

        dates.append(date_text)
        days.append(day_text)
        texts.append(desc_text)
```

```python
# Create dataframe with three columns: date, weekday, text description
df = pd.DataFrame({
    'Date': dates,
    'Day of the Week': days,
    'Text Description': texts
})
```

```python
# Print dataframe
print("\n Fall 2021 Academic Calendar")
print(df.to_string(max_colwidth=75))
```
